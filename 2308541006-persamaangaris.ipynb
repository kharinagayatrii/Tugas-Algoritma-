```python
print("=========================================SELAMAT DATANG AWKHARINERS========================================")
print("==================================Program Penentu Grafik Persamaan Garis===================================")
import numpy as np
import matplotlib.pyplot as plt

print("\nKalian mau nentuin bagaimana bentuk grafik dari persamaan garis yang kalian punya? \nAYO IKUTIN PERMAINAN ATAU RULES PROGRAM INI YA;)")
print("\n")
print("==============================================PENYELESAIAN=================================================")
print("Langkah-langkah untuk menentukkan grafik dari persamaan garis adalah sebagai berikut.")
# Persamaan Garis ini melewati titik koordinat x1,y1 dan x2,y2 di (j_06,k_06) dan (p_06,q_06)
# Input titik pertama (x1,y1)
j_06 = float(input("Masukkan nilai titik koordinat x1: "))
k_06 = float(input("Masukkan nilai titik koordinat y1: "))
# Input titik kedua(x2,y2)
p_06 = float(input("Masukkan nilai titik koordinat x2: "))
q_06 = float(input("Masukkan nilai titik koordinat y2: "))
# Setelah titik koordinatnya diketahui, dilanjutkan menghitung gradien
Gradient_06 = (q_06 - k_06)/(p_06 - j_06)
print("ketemu nih!!!\nNilai Gradien dari Grafik yang akan terbentuk dari persamaan garis kamu adalah {}".format(Gradient_06))
# setelah mendapatkan gradien, subtitusi gradien dan titiknya ke persamaan y_06=Gradient_06*(x_06 - j_06)+k_06 dengan Gradient_06 = gradient
x_06 = np.arange(0,5,0.01) # menentukan batas koordinatnya
plt.figure(figsize=(4,4)) # menentukan besar gambar garisnya
plt.plot(x_06,Gradient_06*(x_06 - j_06)+k_06) # mendefinisikan persamaan garisnya
print ("persamaan garis 1 adalah: y= {}x+{}".format(Gradient_06,(-Gradient_06*j_06)+k_06))
plt.plot(x_06,Gradient_06*(x_06-j_06)+k_06,color="purple",label="y= {}x+{}".format(Gradient_06,(-Gradient_06*j_06)+k_06))
plt.legend(loc= "upper right") # memposisikan dikiri atas agar terlihat rapi
plt.title("Gambar Persamaan Garis") # memberi judul gambar persamaan garisnya
plt.grid(True) # menambahkan garis bantu
plt.xlabel("X") # memberikan label sumbu x
plt.ylabel("Y") # memberikan label sumbu y
plt.show()
print("=============================================TERIMA KASIH================================================")
```

```python
print("=========================================SELAMAT DATANG AWKHARINERS========================================")
print("========================Program Penentu Titik Potong dari 2 Grafik Persamaan Garis==========================")
import numpy as np
import matplotlib.pyplot as plt
# Persamaan Garis Pertama ini melewati titik koordinat x1,y1 dan x2,y2 di (h_06,a_06) dan (r_06,i_06)
# Input titik koordinat garis pertama (x1,y1)
h_06 = float(input("Masukkan nilai titik koordinat x1: "))
a_06 = float(input("Masukkan nilai titik koordinat y1: "))
# Input titik koordinat garis kedua(x2,y2)
r_06 = float(input("Masukkan nilai titik koordinat x2: "))
i_06 = float(input("Masukkan nilai titik koordinat y2: "))

# hitung gradien persamaan garis pertama 
Gradient1_06 =(i_06 - a_06)/(r_06 - h_06)
print ("jika gradien garis pertama adalah: Gradient1 = {}".format(Gradient1_06))
#subtitusi gradien dan titik titik yang diketahui ke y1_06=Gradient1_06*(x_06-h_06)+a_06
print ("maka persamaan garis pertama adalah: y1 = {}x+{}".format(Gradient1_06,-Gradient1_06*h_06+a_06))

# Persamaan Garis Kedua ini melewati titik koordinat x1,y1 dan x2,y2 di (m_06,n_06) dan (s_06,b_06)
# Input titik koordinat garis pertama (x1,y1)
m_06 = float(input("Masukkan nilai titik koordinat x1: "))
n_06 = float(input("Masukkan nilai titik koordinat y1: "))
# Input titik koordinat garis kedua (x2,y2)
s_06 = float(input("Masukkan nilai titik koordinat x2: "))
b_06 = float(input("Masukkan nilai titik koordinat y2: "))

#hitung gradien persamaan garis kedua
Gradient2_06 =(b_06 - n_06)/(s_06 - m_06)
print ("jika gradien garis 2 adalah: Gradient2 = {}".format(Gradient2_06))
#subtitusi gradien dan titik titik yang diketahui ke y2_06=Gradient2_06*(x_06-m_06)+n_06
print ("maka persamaan garis 2 adalah: y2 = {}x+{}".format(Gradient2_06,-Gradient2_06*m_06+n_06))

#kemungkinan yang terjadi apabila terdapat 2 persamaan garis
if Gradient1_06 == Gradient2_06 :
  print ("waahh ternyata gradien dari perdamaan garis kamu keduanya sama, jadi kemungkinan garisnya sejajar atau berhimpit nihh")
else :
  print ("\n, tetapi saat Gradien Garis Pertama tidak sama dengan Gradien Garis Kedua.\nMengartikan 2 garis ini memiliki titik potong")
  #mencari titik potong persamaan garis (TIPO_x,TIPO_y)
  TIPO_x_06 = (Gradient1_06*h_06-a_06 -Gradient2_06*m_06+n_06)/(Gradient1_06-Gradient2_06) #titik potong persamaan garis pertama
  TIPO_y_06 = Gradient2_06*(TIPO_x_06-m_06)+n_06#titik potong persamaan garis kedua
  print ("jadi titik potong kedua garis adalah ({},{})".format(TIPO_x_06,TIPO_y_06)+"\n")

#gambaran hasil grafik persamaan garis beserta titik potongnya
x_06 = np.arange(-10,10,0.01) #tentukan batas koordinat sumbu x
plt.figure(figsize=(4,4)) # menentukan besar gambar garisnya
plt.plot(x_06,Gradient1_06*(x_06-h_06)+a_06,color="blue",label="y1 = {}x+{}".format(Gradient1_06,-Gradient1_06*h_06+a_06))
plt.plot(x_06,Gradient2_06*(x_06-m_06)+n_06,color="red",label="y2 = {}x+{}".format(Gradient2_06,-Gradient2_06*m_06+n_06))
plt.plot(TIPO_x_06,TIPO_y_06,color="black",marker="o",markersize=4) #tandai titik potongnya
plt.title("Gambar Titik Potong Persamaan Garis") #berikan judul gambarnya
plt.grid(True) #tambahkan garis bantu
plt.xlabel("x") #berikan label sumbu x
plt.ylabel("y") #berikan label sumbu y
plt.show()
```
