```python
def hitung_selisih_usia(tahun_anda, tahun_pasangan):
    selisih_usia = abs(tahun_anda - tahun_pasangan)
    return selisih_usia

def main():
    print("Selamat datang di Program Matematika Cinta!")
    tahun_anda = int(input("Masukkan tahun kelahiran Anda: "))
    tahun_pasangan = int(input("Masukkan tahun kelahiran pasangan Anda: "))

    selisih_usia = hitung_selisih_usia(tahun_anda, tahun_pasangan)

    print("Selisih usia Anda dan pasangan Anda adalah", selisih_usia, "tahun.")

    if selisih_usia <= 5:
        print("Wow, Anda dan pasangan Anda memiliki selisih usia yang tidak terlalu jauh!")
        print("Mungkin itu akan membuat hubungan Anda lebih dekat.")
    else:
        print("Selisih usia yang yang cukup berumur, tetapi cinta tidak mengenal batas usia!")
        print("Yang terpenting adalah kesetiaan dan saling pengertian.")

if __name__ == "__main__":
    main()
```
