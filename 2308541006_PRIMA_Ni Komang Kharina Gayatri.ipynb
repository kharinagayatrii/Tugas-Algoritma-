```python
#LATIHAN BILANGAN PRIMA
#Soal:
#Jika diketahui bahwa n adalah dua digit terakhir dari NIM mahasiswa,
#maka tentukan semua Bilangan Prima apa saja yang ada antara interval (1000n, 1000(n+1))
#Diketahui:
#NIM adalah 2308541006, sehingga dua digit terakhir nya adalah 06 artinya n bernilai 6
#Jika n = 6, maka Intervalnya akan bernilai (1000(6), 1000(6+1)) yaitu (6000,7000)
#Penyelesaian:
#Akan ditentukan Bilangan Prima apa saja yang termuat atau terdapat pada interval Bilangan 6000 sampai dengan 7000
```

```python
#Akan diinput 2 digit terakhir NIM
n_06 = int(input("Masukkan 2 digit terakhir dari NIM: "))

#Akan diinput batas bawah dan batas atas dari interval sesuai dengan rumus interval pada soal
Batas_Bawah_Interval06 = 1000 * (n_06) #Batas Bawah Interval
Batas_Atas_Interval06 = 1000 * (n_06 + 1) #Batas Atas Interval
print(f"Bilangan prima yang ada pada interval antara batas bawah yaitu {Batas_Bawah_Interval06} sampai batas atas yaitu {Batas_Atas_Interval06} adalah sebagai berikut: ")

#Akan ditentukkan menggunakan python perulangan (for) dan (if)
#Variabel i memisalkan semua bilangan di rentangan Batas Bawah Interval sampai dengan Batas Atas Interval + 1
#Untuk (perulangan) semua bilangan yang ada pada rentangan Batas Bawah Interval sampai dengan Batas Atas Interval + 1
#Batas Atas Interval ditambahkan 1, karena jika tidak ditambahkan 1 maka batas akhir bilangan yang diinput tidak akan masuk kedalam interval
for i in range(Batas_Bawah_Interval06, Batas_Atas_Interval06 + 1):
  if i > 1: #Saat nilai i lebih kecil atau sama dengan satu yaitu i = 1 dan i = 0 sudah diketahui bukan merupakan Bilangan Prima jadi bisa diabaikan dan terfokus pada nilai i lebih dari 1, karena syarat suatu bilangan dikatakan prima adalah hanya habis dibagi dengan 1 dan habis dibagi oleh dirinya sendiri. Sehingga minimal terdapat 2 faktor
    for n in range(2,i): #Variabel n memisalkan semua bilangan yang ada pada rentangan 2 sampai dengan nilai i, dimana i lebih dari 1
      if (i % n) == 0: #Jika bilangan i mod n adalah 0 atau bilangan i habis dibagi n, maka bilangan tersebut Bukan merupakan Bilangan Prima karena habis dibagi oleh bilangan lain yang bukan 1 atau dirinya sendiri
        break
    else:
      print(i) #Jika bilangan i mod n =! 0 atau bilangan i tidak habis dibagi n, maka bilangan tersebut merupakan Bilangan Prima. Sehingga print hasil bilangan i tersebut.
```
