```python
import numpy as np
print ("!SELAMAT DATANG DI KALKULATOR SEDERHANA MATRIKS!")
print ("================================================")
K_06 = int(input("Masukkan Jumlah Baris Matriks 1 yang kalian mau: "))
A_06 = int(input("Masukkan Jumlah Kolom Matriks 1 yang kalian mau: "))
print("Masukkan Nilai-nilai Matriks 1: ")
entries= list(map(int, input().split()))
matriks1= np.array(entries).reshape(int(K_06),int(A_06))
print("Matriks 1: ")
print(matriks1)

K_06 = int(input("Masukkan Jumlah Baris Matriks 2 yang kalian mau: "))
A_06 = int(input("Masukkan Jumlah Kolom Matriks 2 yang kalian mau: "))
print("Masukkan Nilai-nilai Matriks 2: ")
entries= list(map(int, input().split()))
matriks2= np.array(entries).reshape(int(K_06),int(A_06))
print("Matriks 2: ")
print(matriks2)

print ("\nSilahkan Pilih Perhitungan Matriks Sesuai Angka:\n 1.Penjumlahan Matriks\n 2.Pengurangan Matriks\n 3.Perkalian Matriks")
Pilihan = int(input("Masukkan Pilihan Anda: "))

if Pilihan == 1:
  print("\nHasil Penjumlahan Matriks kalian adalah: ")
  print(matriks1 + matriks2)

elif Pilihan == 2:
  print("\nHasil Pengurangan Matriks kalian adalah: ")
  print(matriks1 - matriks2)

elif Pilihan == 3:
  Perkalian_Matriks06 = np.dot(matriks1, matriks2)
  print("\nHasil Perkalian 2 Matriks kalian adalah: ")
  print(Perkalian_Matriks06)
print ("TERIMA KASIH SUDAH MENGGUNAKAN AKU YAA:))))")
```
