```python
print("===============================================SELAMAT DATANG==============================================")
print("=================================Program Penentu Perpotongan Persamaan Garis===============================")
import numpy as np
import matplotlib.pyplot as plt

print("Kalian mau nentuin perpotongan dari 2 persamaan garis yang kalian punya? AYO IKUTIN PERMAINAN ATAU RULES PROGRAM INI YA;)")
print("\n")
print("=================================PENYELESAIAN====================================")
print("Langkah-langkah untuk menentukkan titik potong antara 2 persamaan garis adalah sebagai berikut.\n1. Tentukan nilai Gradien")
#Persamaan Garis Pertama melewati titik koordinat x1,y1 dan x2,y2 di (j_06,k_06) dan (p_06,q_06)
j_06 = 3 #Titik koordinat (j_06,k_06)
k_06 = 4
p_06 = 2#Titik Koordinat (p_06,q_06)
q_06 = 5
m_1_06 = (q_06-k_06)/(p_06-j_06)
print("ketemu nih!!!\nGradien dari Garis Pertama adalah {}".format(m_1_06))
print("Persamaan Garis Pertama adalah y-{}x+{} ".format(m_1_06,-m_1_06*j_06+k_06))
#y_1_6 - m_1_06*(x-j_06)+k_06
print("\n")
#Persamaan Garis Pertama Kedua melewati titik koordinat x2,y2 di (a_06,b_06) dan (c_06,d_06)
a_06 = 3
b_06 = 4
c_06 = 2
d_06 = -5
m_2_06 = (d_06-b_06)/(c_06-a_06)
print("Gradien dari Garis Kedua adalah {}".format(m_2_06))
print("Persamaan Garis Kedua adalah y={}x+{} ".format(m_2_06, -m_2_06*a_06+b_06))
#y2-m_2_06*(x-a_06)+b_06
print("\n")
if m_1_06 == m_2_06:
    print("Yahhh, Gradienmya sama loh, jadi memiliki 2 kemungkinan yaitu memiliki solusi banyak atau tidak memiliki solusi")
else:
    #Titik Potong Kedua Garis pada (Tp_x06,Tp_y06)
    Tp_x06 = (m_1_06*j_06-k_06-m_2_06*a_06+b_06)/(m_1_06 - m_2_06)
    Tp_y06 = m_2_06*(Tp_x06-a_06)+b_06
    print("Grafik perpotongan dari 2 persamaan garis yang tebentuk adalah sebagai berikut.". format(Tp_x06,Tp_y06))
    
    x=np.arange(Tp_x06-5,Tp_y06,0.01)
    plt.figure(figsize=(4,4),dpi=500)
    
    plt.plot(x,m_1_06*(x-a_06)+b_06,color='blue',label='y-{}x+{}'.format(m_1_06, -m_1_06*j_06+k_06))
    plt.plot(x,m_2_06*(x-a_06)+b_06,color='red',label='y-{}x+{}'. format(m_2_06, -m_2_06*a_06+b_06))
    plt.plot(Tp_x06,Tp_y06,color='green',marker='o',markersize=4)
    plt.legend(loc='best')
    plt.show()
print("\n")
print("===============================================Terima Kasih==============================================")
```
