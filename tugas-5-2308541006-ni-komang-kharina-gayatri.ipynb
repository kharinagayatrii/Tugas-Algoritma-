```python
def Hitung_Depan_06(Miring_06,Samping_06) :
    Depan_06 = sqrt(float(Miring_06**2) - float(Samping_06**2))**(1/2)
    return Depan_06
def Hitung_Samping_06(Miring_06,Depan_06) :
    Samping_06 = sqrt(float(Miring_06**2) - float(Depan_06**2))**(1/2)
    return Samping_06
def Hitung_Miring_06(Depan_06,Samping_06) :
    Miring_06 = sqrt(float(Depan_06**2) + float(Samping_06**2))**(1/2)
    return Miring_06
```

```python
print("SELAMAT DATANG DI PROGRAM PYTHAGORAS")
print("====================================")
Hitung_PYTHAGORAS06 = input("Bagian segitiga mana yang ingin kalian hitung, pilih salah satu ; Depan, Samping, Miring?")
if Hitung_PYTHAGORAS06.upper() == 'DEPAN' :
    Miring_06 = float(input("Berapakah nilai miring segitiganya?"))
    Samping_06 = float(input("Berapakah nilai samping segitiganya?"))
    Depan_06 = (float(Miring_06**2) - float(Samping_06**2))**(1/2)
    print(f"Nilai {Hitung_PYTHAGORAS06} segitiga siku-siku adalah " + str(Depan_06))
elif Hitung_PYTHAGORAS06.upper() == 'SAMPING' :
    Miring_06 = float(input ("Berapakah nilai miring segitiganya?"))
    Depan_06 = float(input("Berapakah nilai depan segitiganya?"))
    Samping_06 = (float(Miring_06**2) - float(Depan_06**2))**(1/2)
    print(f"Nilai {Hitung_PYTHAGORAS06} segitiga siku-siku adalah {Samping_06}")
elif Hitung_PYTHAGORAS06.upper() == 'MIRING' :
    Depan_06 = float(input("Berapakah nilai depan segitiganya?"))
    Samping_06 = float(input("Berapakah nilai samping segitiganya?"))
    Miring_06 = (float(Depan_06**2) + float(Samping_06**2))**(1/2)
    print(f"Nilai {Hitung_PYTHAGORAS06} segitiga siku-siku adalah {Miring_06}")
```
